```python
def DSL_sur_SD_piC():
    DSL_sur_SD_pic=[]
    for t in range(210):
        calcul = (ZOS_mean[t]-zos_piC_yearmean)/SD_piC_DSL
        DSL_sur_SD_pic.append(calcul)
    return DSL_sur_SD_pic
```

```python
DSL_sur_SD_mean = DSL_sur_SD_piC()
```

```python
def GMTSL_sur_SD_piC(i):
    GMTSL_sur_SD_pic=[]
    for t in range(130,210):
        calcul = (GMTSL[i][t]-GMTSL_piC_yearmean)/SD_piC_DSL
        GMTSL_sur_SD_pic.append(calcul)
    return GMTSL_sur_SD_pic
```

```python
GMTSL_sur_SD =[GMTSL_sur_SD_piC(i) for i in range(31)]
```

```python
def ToE_DSL_pos():
    ToE_DSL = zos_piC[0].copy()
    
    for (y,x) in indice_ocean :
            t=209        
            while (float(DSL_sur_SD_mean[t][y,x]))>=2 and t>0:
                t =t-1
         
            C = 209 - (t+1)
        
            if C>=20:
                ToE_DSL[y,x]=1850+(t+1)
        
            else:
                ToE_DSL[y,x]=np.nan
            
            
    return ToE_DSL
```

```python
def ToE_DSL_neg():
    ToE_DSL = zos_piC[0].copy()
    
    for (y,x) in indice_ocean :
            t=209        
            while (float(DSL_sur_SD_mean[t][y,x]))<=-2 and t>0:
                t =t-1
         
            C = 209 - (t+1)
        
            if C>=20:
                ToE_DSL[y,x]=1850+(t+1)
        
            else:
                ToE_DSL[y,x]=np.nan
            
            
    return ToE_DSL
```

```python
# Median & 16-84% range en tenant compte de positif et de négatif comme dans Lyu et al.

# Contraintes :
# Il faut que plus de la moitié des modèles donne soit la même direction soit pas d'emergence
# Si on a plus de 1 modèle qui donne direction opposée : Non

#on définit une fonction de i, le nombre de modèle qu'on considère

def percentile_ToE(i):
    
    median_pos = ToE_GMTSL_r1_CC.copy()
    median_neg = ToE_GMTSL_r1_CC.copy()
    range_16_84 = ToE_GMTSL_r1_CC.copy()
    
    for (y,x) in indice_ocean:

        C_pos=i    
        L_pos=[k for k in range(i)]
        C_nan=0
        C_neg=0
        L_neg =[]
        
        for j in range(i):
            
            if np.isnan(ToE_pos[j][y,x]) and np.isnan(ToE_neg[j][y,x]):
                C_nan=C_nan+1
                L_pos.remove(j)
                C_pos=C_pos-1
                
            if  ToE_neg[j][y,x]>0:
                C_neg=C_neg+1
                L_neg.append(j)
                L_pos.remove(j)
                C_pos=C_pos-1
            
        if C_nan > i/2:
                
            median_pos[y,x]=np.nan
            median_neg[y,x]=np.nan
            range_16_84[y,x]=np.nan
                
        elif C_pos >= i/2 and C_neg<=1:
                
            L=[ToE_pos[j][y,x] for j in L_pos]                
            Quart = np.percentile(L, [16, 50, 84])
                
            median_pos[y,x]=Quart[1]
            median_neg[y,x]=np.nan
            range_16_84[y,x]=Quart[2]-Quart[0]
            
        elif C_neg >=i/2 and C_pos<=1:
                
            L=[ToE_neg[j][y,x] for j in L_neg]
            Quart = np.percentile(L, [16, 50, 84])
                
            median_neg[y,x]=Quart[1]
            median_pos[y,x]=np.nan
            range_16_84[y,x]=Quart[2]-Quart[0]
        
        else : 
                
            median_pos[y,x]=np.nan
            median_neg[y,x]=np.nan
                
            range_16_84[y,x]=np.nan

    return median_pos, median_neg, range_16_84
    
```

```python
median_ToE_pos,median_ToE_neg, range_16_84 = percentile_ToE(31)
```
